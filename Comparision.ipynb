# **Comparision Of 3 Languages**

```python
import matplotlib.pyplot as plt

size = [200, 400, 600, 800, 1000]

python_time = [1.81, 4.74, 7.90, 11.36, 13.05]
c_time = [2.09, 3.82, 6.58, 8.93, 11.20]
cpp_time = [2.28, 4.25, 6.93, 9.49, 12.08]

plt.figure(figsize=(10, 6))
plt.plot(size, c_time, marker='o', label='C')
plt.plot(size, cpp_time, marker='s', label='C++')
plt.plot(size, python_time, marker='x', label='Python')

plt.xlabel("File Size (MB)")
plt.ylabel("Time Taken (seconds)")
plt.title("File Processing Time vs File Size")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
